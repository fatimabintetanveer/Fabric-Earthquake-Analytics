```python
import requests
import json

# Define the URL for the API request using the start and end dates
url = f"https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime={start_date}&endtime={end_date}"

# Send a GET request to the API
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response and extract the 'features' section
    data = response.json()
    data = data['features']

    # Define the file path for saving the data
    file_path = f'/lakehouse/default/Files/{start_date}_earthquake_data.json'

    # Save the extracted data to a JSON file
    with open(file_path,'w') as file:
        json.dump(data, file, indent=4)  # Write JSON data with indentation
        print("Data successfully saved!")  # Confirmation message

else:
    # Print an error message if the API request failed
    print("Failed to fetch the data. Status code: ",response.status_code)
```
